```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5n')
```

```python
model
```

```python
img = 'image.jpg'
results = model(img)
results.print()
```

```python
resized = cv2.resize(img, dsize=(240, 680), interpolation=cv2.INTER_CUBIC)
results_resized = model(resized)
results.print()
```

```python
%matplotlib inline 
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture('../testVideos/vid5.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
